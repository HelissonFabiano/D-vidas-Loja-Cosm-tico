Porém, nem todas as tags 'h2' armazenam os valores dos boletos. Filtrando os valores de 'h2' e percorrendo cada um desses valores, chegamos a conclusão que os 31 primeiros elementos 'h2' armazenam valores referentes aos boletos.

Chegamos a conclusão de que se extrairmos os 31 primeiros valores de 'h2', obteremos os valores individuais de cada boleto.

Para isso, vamos utilizar a biblioteca bs4(Beautiful Soup) do Python, que nos possibilita analisar e iterar sobre arquivos html e xml.





Primeiramente, será necessário importar o arquivo html da página de pedidos da revendedora. No nosso caso, foi feito o download do arquivo em formato html e, posteriormente, o arquivo foi upado no Google Colab para ser utilizado.

```python
from bs4 import BeautifulSoup #Importando a biblioteca BS4 para iterar sobre o arquivo html
```

```python
#Lendo o conteúdo do arquivo e armazenando em uma variável

with open("/content/Revendedor O Boticário.html", "r", encoding="utf-8") as arquivo:
    conteudo = arquivo.read()
```

```python
#criando um objeto que representa o arquivo html e armazenando em uma variável
soup = BeautifulSoup(conteudo, "html.parser")
```

```python
#buscando todos os valores referentes a tag 'h2' e armazenando em uma variável
dividas = soup.find_all('h2')
```

```python
total_dividas = [] #Lista que armazenará somente os 31 primeiros valores de 'h2'

#armazenando os 31 primeiros valores de 'h2' em uma lista e tratando os dados para transformá-los no tipo float,
#permitindo que operações matemáticas sejam realizadas com estes dados.
for i in range(0,31):

  total_dividas.append((dividas[i]).text[3:])
  total_dividas[i] = float(total_dividas[i].replace(',', '.'))

```

Agora, se mostramos o valor de 'total_dividas', será mostrada uma lista com os valores dos boletos, e cada elemento dessa lista será do tipo 'float', o que permite que esses valores sejam iterados.

```python
total_dividas
```

Vamos utilizar a função sum() para somar os valores de 'total_dividas', alcançando o objetivo do nosso projeto , que é apenas somar o valor das dívidas da loja.

```python
divida_total = sum(total_dividas)

print(f'O valor total da dívida é de R$ {divida_total}')
```
